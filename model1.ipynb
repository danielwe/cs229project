# Learning to move a human arm

This notebook runs several instances of the script `model1.py`, and gathers and plots the results

```python
from multiprocessing import Pool
from model1 import main, file_prefix, pickleload

# Set parameters to sweep
sigmas = [0.05, 0.1, 0.15, 0.2]
learning_rates = [0.0001, 0.001, 0.01]
```

```python
p = Pool(processes=12, maxtasksperchild=1)
p.imap_unordered(main, [(s, lr, file_prefix(s, lr))
                        for s in sigmas
                        for lr in learning_rates])
```

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
from plots import (plot_observations,
                   plot_reward_history,
                   plot_reward_statistics)
```

```python
training_history = {(s, lr): pickleload(file_prefix(s, lr) + '_training_history.pkl')
                    for s in sigmas for lr in learning_rates} 
test_history = {(s, lr): pickleload(file_prefix(s, lr) + '_test_history.pkl')
                for s in sigmas for lr in learning_rates} 
```

```python
# Plot training reward history
def latex_s(s):
    return r'$\sigma = {}$'.format(s)

def latex_lr(lr):
    return r'$\alpha = {}$'.format(lr)

def latex(s, lr):
    return latex_s(s)[:-1] + ', ' + latex_lr(lr)[1:]

for lr in learning_rates:
    axes = None
    for s in sigmas:
        history = training_history[(s, lr)]
        axes = plot_reward_history(history, axes=axes, label=latex_s(s))
    axes.legend(loc=0)
    axes.figure.savefig('model1_sigma-{}_training_rewards.pdf'.format(s))
```

```python
# Plot test observation history
for key, history in test_history.items():
    axes = plot_observations(history)
    axes[0].figure.savefig(file_prefix(*key) + '_test_observations.pdf')
    print(key)
```

```python
from matplotlib import pyplot as plt
for lr in learning_rates:
    fig, axes = plt.subplots()
    plot_reward_statistics({latex_s(s): test_history[(s, lr)] for s in sigmas}, axes=axes)
    axes.figure.savefig('model1_lr-{}_test_rewards.pdf'.format(lr))
```
