```python
%matplotlib inline
from matplotlib import pyplot as plt
from rl.random import OrnsteinUhlenbeckProcess
```

```python
for sigma in 0.2, 0.15, 0.10, 0.05:
    process = OrnsteinUhlenbeckProcess(theta=.15, mu=0., sigma=sigma, dt=0.01)
    samples = [process.sample() for _ in range(50000)]
    plt.plot(samples, alpha=0.75, label=r'$\sigma = {}$'.format(sigma))
plt.xlabel('t$\ / \ \mathrm{steps}$')
plt.ylabel('noise')
plt.legend(loc=0)
plt.savefig('noise.pdf')
```
